```python
#basics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#seaborn can load dataset
a = sns.load_dataset("flights")
sns.relplot(x ="passengers", y ="month", data =a)
```

```python
#hue helps in categorizing data according to given data
a = sns.load_dataset("flights")
sns.relplot(x ="passengers", y ="month",hue="year", data =a)
```

```python
#relplot can helps to plot line plot also
b = sns.load_dataset("tips")
sns.relplot(x ="time", y ="tip", data =b, kind="line")
```

```python
#plotting with categorical - use catplot
sns.catplot(x="day", y="total_bill", data =b)
```

```python
sns.catplot(x="day", y="total_bill", data =b, kind="violin")
```

```python
sns.catplot(x="day", y="total_bill", data =b, kind ="boxen")
```

```python
#plot univariate
c = np.random.normal(loc =5, size =100, scale =2) #pick random variable which has normal distribution
sns.distplot(c)
```

```python
#plot multi plot grids -  can use facet grid and pair grid
d = sns.load_dataset("iris")
e = sns.FacetGrid(d, col ="species")
e.map(plt.hist, "sepal_length")
```

```python
a= sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
#plot - Aesthetics
sns.set(style="dark")
a= sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
#boxplot
sns.set(style="white", color_codes=True)
a = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=a)
```

```python
#remove axes line
sns.set(style="white", color_codes=True)
a = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=a)
sns.despine(offset =10, trim =True)
```
